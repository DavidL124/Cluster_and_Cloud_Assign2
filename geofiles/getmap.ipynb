```python
def convert_centroid_to_point(tweet):
    ret = []
    tweet_spt = tweet.split(",")
    ret[0] = tweet_spt[0][1:]
    ret[1] = tweet_spt[1][1:-1]
    return ret
```

```python
def get_coords(tmp):    
    with open("twitter_stream_group_geo.json", "r") as f:
        for line in f.readlines():
            print("line = ", line[:5])
            if line[:5] == '{"id"':
                print(line)
                if line[-2] == ',':
                    line = line[:-2]
                elif line[-2] != ',' and line[-3] == ']':
                    line = line[:-3]
                tweet = json.loads(line)
                if "geo" in tweet.keys() and tweet["geo"] != {}:
                    #If the coordinates attribute is available we use that and the centroid.
                    if "coordinates" in tweet["geo"].keys():
                        tmp.append(Point(tweet["geo"]["coordinates"]))
                    elif "coordinates" not in tweet["geo"].keys():
                        tmp.append(Point(convert_centroid_to_point(tweet["geo"]["geo_location"]["centroid"])))
                    
```

```python
geo = [[-0.009312629699707031, 51.54724243973627], [-80.61101417488993, 28.331331499999997], 
       [144.91196999, -37.782276],[126.90356200667841, 37.5558529140736], [144.91196999,-37.782276], 
      [145.15252629070534, -37.9725599375], [144.9661, -37.8156], [144.2845306, -36.757164]]
```

```python
geo_points = []
for item in geo:
    geo_points.append(Point(item[0], item[1]))
```

```python
#Plot the points on the map:
def within_file(point, shapes):
    for shape in shapes:
        if shape != None:
            if shape.contains(point) or shape.touches(point):
                #Point contained
                return True
    return False
```

```python
#First try out geojson:
#https://maximeborry.com/post/interactive-map-howto/
australia = []
lat = []
long = []
marker = []
for point in geo_points:
    if within_file(point, shapefile.geometry):
        print("In Australia")
        australia.append(point)
        lat.append(point.x)
        long.append(point.y)
        marker.append("#009C54")
        
```
